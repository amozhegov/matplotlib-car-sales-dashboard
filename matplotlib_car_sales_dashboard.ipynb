```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Evenly spaced numbers over the interval [0, 10]
x = np.linspace(0, 10, 1000)
```

```python
y1 = np.power(x, 0.5)
plt.plot(x, y1, 'b--', label='$\sqrt{x}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Square root function')
plt.legend()
```

```python
y2 = 3 * np.sin(x)
plt.plot(x, y2, 'go', label='$3 \dot \sin{x}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Sine function')
plt.legend()
```

```python
y3 = 2 * np.cos(x)
plt.plot(x, y3, 'r:', label='$2 \dot \cos{x}$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Cosine function')
plt.legend()
```

```python
# Subplots
fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize=[14,3], dpi=400)
ax1, ax2, ax3 = axes
ax1.plot(x, y1, 'b--')
ax2.plot(x, y2, 'go')
ax3.plot(x, y3, 'r:')
fig.suptitle('Square root, Sine and Cosine functions')
```

```python
names = ['BMW', 'Volkswagen', 'Toyota']
values = [100, 150, 180]

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10), dpi=400)
(ax1, ax2), (ax3, ax4) = axes

# Pie Chart
ax1.pie(values, colors=['red', 'green', 'blue'], 
        autopct='%1.1f%%', 
        hatch=['**O', 'oO', 'O.O'], 
        textprops={'color':"w", 'weight':'bold', 'fontsize':15.5})
ax1.set_title('Auto Sales Pie Chart')
ax1.legend(names, title = 'Auto Sales',bbox_to_anchor=(-0.25, 0, 0.5, 1))

# Bar Plot
ax2.bar(names, values, color=['r', 'g', 'b'], 
        edgecolor='black', bottom=int(values[0])-20,
        hatch=['**O', 'oO', 'O.O'])
ax2.set_title('Auto Sales Bar Plot')
plt.xlabel('Companies')
plt.ylabel('Sales')

# Stem plot
ax3.stem(names, values, linefmt='b:', orientation='horizontal')
ax3.set_title('Auto Sales Plot')

# Stackplot
companies = ['BMW', 'Volkswagen', 'Toyota']
values_1 = [120, 100, 80]
values_2 = [100, 125, 165]
values_3 = [145, 185, 165]
colors = ['lightcoral', 'lightgreen', 'skyblue']
plt.stackplot(companies, values_1, values_2, values_3, labels=['BMW', 'Volkswagen', 'Toyota'], colors=colors)
plt.xlabel('Categories')
plt.ylabel('Sales, ')
plt.title('Stackplot Auto Sales')
plt.legend(loc='upper left')
plt.show()
```
